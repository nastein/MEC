```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

```python
pp_nn = pd.read_csv('/Users/noah/Physics/Fermilab/MEC/code/C12_2700_15p0_2b_pp+nn.out',sep="\s+|\t+|\s+\t+|\t+\s+", header=None, engine='python')
pn_np = pd.read_csv('/Users/noah/Physics/Fermilab/MEC/code/C12_2700_15p0_2b_pn+np.out',sep="\s+|\t+|\s+\t+|\t+\s+", header=None, engine='python')
total = pd.read_csv('/Users/noah/Physics/Fermilab/MEC/code/C12_2700_15p0_2b_FSI2_harry_full.out',sep="\s+|\t+|\s+\t+|\t+\s+", header=None, engine='python')

pp_nn.columns = ['w','Emu','xsec','xsec_err']
pn_np.columns = ['w','Emu','xsec','xsec_err']
total.columns = ['w','Emu','xsec','xsec_err']
```

```python
fig1, ax = plt.subplots(1,1)
plt.tight_layout(pad=1.0)

ax.plot(pp_nn['w'],pp_nn['xsec'],label="pp+nn initial",color='red')
ax.plot(pn_np['w'],pn_np['xsec']*2,label="pn+np initial",color='blue')
ax.plot(pn_np['w'],pn_np['xsec']*2 + pp_nn['xsec'],label="pp+nn+pn+np",color='green')
ax.plot(total['w'],total['xsec'],label="summed over initial",color='black')
ax.set_xlabel(r'$\omega$ (MeV)')
ax.set_ylabel(r'$d\sigma/d\Omega d\omega$ (nb/MeV/sr)')
ax.set_title(r'e-C12 Scattering, $E_{e} = 1500$ MeV, $\theta_{e} = 30^{\circ}$')
ax.legend()
```
