```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

```python
#Grab prediction
split_dir='/Users/noah/Physics/Fermilab/MEC/code/'
summed_dir='/Users/noah/Physics/Fermilab/MEC/orig_code/'


summed =summed_dir+f'C12_2700_15p0_2b_FSI2_harry_full.out'
ppnn = split_dir+f'C12_2700_15p0_2b_pp+nn.out'
pnnp = split_dir+f'C12_2700_15p0_2b_pn+np.out'

summed_df = pd.read_csv(summed,sep="\s+|\t+|\s+\t+|\t+\s+", header=None, engine='python')
ppnn_df = pd.read_csv(ppnn,sep="\s+|\t+|\s+\t+|\t+\s+", header=None, engine='python')
pnnp_df = pd.read_csv(pnnp,sep="\s+|\t+|\s+\t+|\t+\s+", header=None, engine='python')

summed_df.columns = ['w', 'Emu', 'xsec','xsec_err']
ppnn_df.columns = ['w', 'Emu', 'xsec','xsec_err']
pnnp_df.columns = ['w', 'Emu', 'xsec','xsec_err']
```

```python

fig1, ax = plt.subplots(1,1,figsize=(5, 5))
#fig1, ax = plt.subplots()


ax.plot(summed_df['w'],summed_df['xsec'], label=r'Summed over all states',color='red')
ax.plot(ppnn_df['w'], ppnn_df['xsec'], label=r'pp+nn initial states', color='blue')
ax.plot(pnnp_df['w'], pnnp_df['xsec']*2.0, label=r'pn+np initial states', color='green')
ax.plot(pnnp_df['w'], (pnnp_df['xsec']*2.0)+ppnn_df['xsec'], label=r'Sum of pp,nn,np,pn states', color='black')
        
ax.set_title(r'MEC contribution for $E_{e} = 1.5$ GeV $\theta_{l} = 30.0$ degrees')
ax.set_xlabel(r'$\omega$ (MeV)', fontsize=11)
ax.set_ylabel(r'$d\sigma^{MEC} /d\Omega d\omega$ (nb/MeV/sr)', fontsize=11)
     #   #ax[i][j].set_yscale('log')
      #  ax[i][j].set_xbound(0,1000)
ax.legend()

#ax.set_ybound(1e-42, 3e-41)

```
